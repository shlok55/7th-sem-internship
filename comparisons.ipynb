```python
import matplotlib.pyplot as plt

# List of model names and their corresponding accuracy scores
model_names = ['KNN', 'Randomforest', 'Decisiontree', 'SVM', 'NaiveBayes']
accuracy_scores = [78.75, 64.25, 84.46115288220551, 73.75, 78.0]

# Scale the accuracy scores to a range of 0 to 100
accuracy_scores_scaled = [score * 100 for score in accuracy_scores]

# Plot the graph
plt.plot(model_names, accuracy_scores_scaled, marker='o', linestyle='-', color='blue')

plt.xlabel('Model')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison')
plt.ylim(0, 100)  # Set the y-axis limits to range from 0 to 100

# Display the accuracy scores on the data points
for i in range(len(model_names)):
    plt.text(model_names[i], accuracy_scores_scaled[i], str(round(accuracy_scores_scaled[i], 2)),
             ha='center', va='bottom', fontsize=8)

plt.show()
```

```python
import chartify

# List of model names and their corresponding accuracy scores
model_names = ['KNN', 'Randomforest', 'Decisiontree', 'SVM', 'NaiveBayes']
accuracy_scores = [78.75, 64.25, 84.46115288220551, 73.75, 78.0]

# Create a DataFrame with the data
data = chartify.DataFrameColumnChart(x=model_names, y=accuracy_scores)

# Create a bar chart
chart = chartify.Chart(blank_labels=True)
chart.plot.bar(data, 'x', 'y')

chart.axes.set_xlabel('Model')
chart.axes.set_ylabel('Accuracy')
chart.axes.set_ylim(0, 100)  # Set the y-axis limits to range from 0 to 100

chart.show()
```
